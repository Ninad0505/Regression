```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas_datareader as pdr
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
tickers = ['^NSEI']

start = dt.datetime(2009, 1, 1)

data = pdr.get_data_yahoo(tickers, start)

data.head(10)
```

```python
log_returns = pd.DataFrame()
```

```python
data['Gap1'] = np.log((data['Open'] - data['Close'].shift())/data['Close'].shift())
data['Gap2'] = (data['Open'] - data['Close'].shift())/data['Close'].shift()
```

```python
data[['Open', 'Close', 'Gap1', 'Gap2']].head(10)
#clear signs that log wont help because on some days there is no gap
```

```python
data['prev_day_bar'] = data['Close'].shift() - data['Open'].shift() 
```

```python
data['prev_day_bar%'] = ( data['Close'].shift() - data['Open'].shift() )/ data['Open'].shift()
```

```python
data = data.drop(['prev_day_bar'], axis=1)
```

```python
data.head()
```

```python
data = data.drop(['Gap1'], axis=1)
data.head()
```

```python
data = data.drop(['Volume'], axis=1)
data.head()
```

```python
data = data.drop(['Adj Close'], axis=1)
data.head()
```

```python
data['prev_day1%'] = ( data['Close'].shift() - data['Close'].shift(2) )/ data['Close'].shift(2)
data['prev_day2%'] = ( data['Close'].shift(2) - data['Close'].shift(3) )/ data['Close'].shift(3)
data['prev_day3%'] = ( data['Close'].shift(3) - data['Close'].shift(4) )/ data['Close'].shift(4)
```

```python
data.head(10)
```

```python
data['net_change'] = data['Open'] - data['Close']
data['net_change_sign'] = [ 1 if i >= 0 else 0 for i in data.net_change ]
data.head()
```

```python
df = data.dropna()
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Gap2','prev_day1%','prev_day2%','prev_day3%']],df.net_change_sign,train_size=0.85)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
df_model = pd.DataFrame()
df_model['Actual'] = y_test
df_model['Predicted'] = y_predicted

df_model.head()
```

```python
df_model['dif'] = abs(df_model['Actual'] - df_model['Predicted'])
```

```python
df_model[['Probability', 'Anti']] = model.predict_proba(X_test)
```
