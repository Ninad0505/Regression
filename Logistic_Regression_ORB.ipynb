```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas_datareader as pdr
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
tickers = ['^NSEI']

start = dt.datetime(2009, 1, 1)

data = pdr.get_data_yahoo(tickers, start)

data.head(10)
```

```python
data = data.drop(['Volume'], axis=1)
data = data.drop(['Adj Close'], axis=1)
data.head()
```

```python
data['Gap'] = data['Open'] - data['Close'].shift() 
data['prev_day1%'] =  data['Close'].shift() - data['Close'].shift(2) 
data['prev_day2%'] =  data['Close'].shift(2) - data['Close'].shift(3) 
data['prev_day3%'] =  data['Close'].shift(3) - data['Close'].shift(4) 
data.head()
```

```python
data['net_change'] = data['Open'] - data['Close']
data['target'] = [ 1 if i >= 0 else 0 for i in data.net_change ]
data['prev_day1%_sign'] = [ 1 if i >= 0 else 0 for i in data['prev_day1%'] ]
data['prev_day2%_sign'] = [ 1 if i >= 0 else 0 for i in data['prev_day2%'] ]
data['prev_day3%_sign'] = [ 1 if i >= 0 else 0 for i in data['prev_day3%'] ]
data.head()
```

```python
data['Gap_sign'] = [ 1 if i >= 0 else 0 for i in data['Gap'] ]
data.head()
```

```python
df = data.dropna()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Gap_sign','prev_day1%_sign','prev_day2%_sign','prev_day3%_sign']],df.target,train_size=0.85)
```

```python
model = LogisticRegression(class_weight = 'balanced')
```

```python
model.fit(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
df_model = pd.DataFrame()
df_model['Actual'] = y_test
df_model['Predicted'] = y_predicted

df_model.head(10)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
cm7 = confusion_matrix(y_test, y_predicted)
print(cm7)
```
